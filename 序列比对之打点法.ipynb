# 比较两个序列的方法
## 打点法
将序列1，2分别写在坐标轴上，再在每个位置上，相同的就打个点。最后，取最近的线。

```python
#py中一个用于绘图的工具
import matplotlib.pyplot as plt
```

```python
#将X和Y中的元素遍历，若相同，则记录下是X第i个与Y的第j个元素相同。
def dot_align (X,Y):
    #中文显示
    plt.rcParams['font.sans-serif'] = ["simhei"]

    plt.title("打点法比对序列")
    plt.xlabel('seq1')
    plt.ylabel('seq2')

    #将坐标轴数字与序列文字一一对应并替换数字
    plt.xticks(list(range(len(X))),X)
    plt.yticks(list(range(len(Y))),Y)

    #坐标轴刻度从-1开始，长度为序列长度
    plt.axis([-1,len(X),-1,len(Y)])
    
    
    i = 0
    a = []
    b = []
    
    c = []
    d = []

    for x in X:
        j = 0
        for y in Y:
            if x == y :
                a.append(i)
                b.append(j)
                #print('计算中。。。')
                #plt.plot(i,j,color='green', marker='o', linestyle='none')
            else:
                #plt.plot(i,j,color='red', marker='o', linestyle='none')
                c.append(i)
                d.append(j)
            
            j = j + 1
        i = i +1
    




    plt.plot(a,b,color='red', marker='o', linestyle='none')
    plt.plot(c,d,color='black', marker='x', linestyle='none')
    #plt.show()
```

```python
#X = "thefastcat"
#Y = "thefatcat"
#输入两个要比对的序列,序列1和2 分别记录为字符串X，Y
#将输入的序列统一为大写，并去除空格
X = input("please input seq1: ").upper().replace(' ', '')
Y = input("please input seq2: ").upper().replace(' ', '')
#print('seq1 is ',X)
#print('seq2 is ',Y)
dot_align(X,Y)
```

两条不同的序列打点，可以看到序列相似的部分。序列与自身打点，可以看到自身的重复片段

```python
example="themotherfuckeristhemotherfucker"
dot_align(example,example)
plt.title("打点法比对序列示例-motherfucker")
```
